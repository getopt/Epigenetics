```r
# Function to run a single experiment
Run_experiment <- function(size, ns_mean, ns_sd, cs_mean, cs_sd)
{
    ns_data <- rnorm(size, mean = ns_mean, sd = ns_sd)
    cs_data <- rnorm(size, mean = cs_mean, sd = cs_sd)

    test <- t.test(ns_data, cs_data)
    return(test$p.value)
}
```

```r
# Function to kick off many individual experiments
Run_many_experiments <- function(max_group_size, n_expr, ns_mean, ns_sd, cs_mean, cs_sd)
{
    all_results <- list()
    for (size in c(2:max_group_size))
    {
        results <- NULL
        for (experiment in c(1:n_expr))
        {
            results <- c(results, Run_experiment(size, ns_mean, ns_sd, cs_mean, cs_sd))
        }

        all_results[[size]] <- -log10(results)
    }
    return(all_results)
    
}
```

```r
# plot
Plotting <- function(all_results, cpgName, nexpr, max_group_size, ns_mean, ns_sd, cs_mean, cs_sd)
{

    # set dimensions of the plot area
    library(repr)
    options(repr.plot.width = 10, repr.plot.heigh = 6)

    boxplot(all_results,
            ylab = "-log10(p.value)", 
            xlab = "Sizes of groups of participants to compare",
            range = 0.5,
            main = paste("Outcomes of ", nexpr, " experiments comparing ", cpgName, " methylation \n",
                         "between non-smokers and current smokers\n",
                         "in groups from 2 to ", max_group_size, " participants"))
    abline(h = -log10(0.05), col = "darkgreen", lwd = 2)
    abline(h = -log10(0.2), col = "red", lwd = 2)
    legend("topleft", pch = 19, col = "white",
           legend = c(paste("Never-smokers mean methylation: ", 100*ns_mean, "%", sep = ""),
                      paste("Never-smokers standard deviation: ", ns_sd, sep = ""),
                      paste("Current smokers mean methylation: ", 100*cs_mean, "%", sep = ""),
                      paste("Current smokers standard deviation: ", cs_sd, sep = "")),
           bty = "n")

}
```

```r
# STAGE EXPERIMENTS

# CpG name
CPG_NAME <- "cg03636183"

# Never smokers (ns): Mean and standard deviation of beta
NS_MEAN <- 0.672
NS_SD <- 0.032

# Current smokers (cs): Mean and standard deviation of beta 
CS_MEAN <- 0.553
CS_SD <- 0.076

# Number of experiments and group size to compare
N_EXPR <- 100
MAX_GROUP_SIZE <- 20
```

```r
# Run experiments

ALL_RESULTS <- Run_many_experiments(MAX_GROUP_SIZE, N_EXPR, NS_MEAN, NS_SD, CS_MEAN, CS_SD)
Plotting(ALL_RESULTS, CPG_NAME, N_EXPR, MAX_GROUP_SIZE, NS_MEAN, NS_SD, CS_MEAN, CS_SD)
```
